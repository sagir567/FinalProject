```python
import os
def process_files(directory):
    """
    Process files in the given directory.
    
    Args:
    - directory (str): Path to the directory containing the files.
    
    Returns:
    - dict: A dictionary mapping file indices to file paths.
    - dict: A dictionary mapping file indices to file text.
    """
    file_paths = {}
    file_texts = {}

    # Iterate over subdirectories in the directory
    for subdirectory in os.listdir(directory):
        subdirectory_path = os.path.join(directory, subdirectory)
        if os.path.isdir(subdirectory_path):
            # Iterate over files in the subdirectory
            for index, filename in enumerate(os.listdir(subdirectory_path)):
                filepath = os.path.join(subdirectory_path, filename)
                if os.path.isfile(filepath):
                    # Store file index and path
                    file_paths[index] = filepath
                    
                    # Read file contents and store them
                    with open(filepath, 'r') as file:
                        file_texts[index] = file.read()
    
    return file_paths, file_texts

# Example usage:
directory = "Desktop/BS'c ComputerScience/thirdYear/finalProject/wave2vec2emotionRecognitionDataSets/TIMIT/data/TEST'
file_paths, file_texts = process_files(directory)



```
